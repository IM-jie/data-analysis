# F分布

## 基本概念

- **F分布**是一种连续型概率分布，常用于比较两个样本的方差，比如：
    - 方差齐性检验
    - 单因素方差分析（ANOVA）
- F分布是两个独立卡方分布的比值：
$$F = \frac{(X_1/d_1)}{(X_2/d_2)}$$
其中：
- $X_1 \sim \chi^2(d_1)$：自由度$d_1$
- $X_2 \sim \chi^2(d_2)$：自由度$d_2$

## 概率密度函数

对于$F \sim \text{F}(d_1, d_2)$的随机变量，其概率密度函数为：
$$f(x) = \frac{\sqrt{\left( \frac{d_1 x}{d_1 x + d_2} \right)^{d_1} \left( \frac{d_2}{d_1 x + d_2} \right)^{d_2}}}{x \cdot B\left(\frac{d_1}{2}, \frac{d_2}{2} \right)}, \quad x > 0$$
其中：
- $B(\cdot, \cdot)$：贝塔函数 
- $d_1, d_2$：自由度参数

## 均值与方差

**均值：**

$$
\text{E}(F) = \frac{d_2}{d_2 - 2}, \quad d_2 > 2
$$

**方差：**

$$
\text{Var}(F) = \frac{2d_2^2(d_1 + d_2 - 2)}{d_1(d_2 - 2)^2(d_2 - 4)}, \quad d_2 > 4
$$


## 性质

- **非负性**：F分布的定义阈值是$x > 0$
- **偏态分布**：随着自由度的变化形状发生改变，尤其是右偏 
- **右尾重**：极端大值概率相对较高 
- **单峰性**：具有一个峰值（mode） 
- **自由度对称性**：
$$\frac{1}{F(d_1, d_2)} \sim F(d_2, d_1)$$

## 图形

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f
import seaborn as sns
sns.set(style="darkgrid")

# 自由度组合
dfs = [(1, 1), (5, 2), (5, 5), (10, 10), (20, 20)]

# x 值范围
x = np.linspace(0.01, 5, 500)

# 绘图
plt.figure(figsize=(10, 6))
for d1, d2 in dfs:
    y = f.pdf(x, d1, d2)
    plt.plot(x, y, linewidth=2, label=f'F(d1={d1}, d2={d2})')

plt.title('F Distributions with Different Degrees of Freedom', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend(title='Distributions', fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

# 与其他分布的关系

| 关系链         | 描述                                                           |
| ----------- | ------------------------------------------------------------ |
| 正态分布 → 卡方分布 | 若 $Z_i \sim N(0, 1)$，则 $\sum Z_i^2 \sim \chi^2(k)$           |
| 卡方分布 → F 分布 | 若 $\frac{U_1/d_1}{U_2/d_2}$，其中 $U_1, U_2 \sim \chi^2$，则 F 分布 |
| F 分布 → 正态近似 | 自由度越大时，F 分布的形状接近正态分布（**中心极限定理**）                                         |
