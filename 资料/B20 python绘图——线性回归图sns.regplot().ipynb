# 线性回归图

**线性回归图（Linear Regression Plot）**是一种用来展示数据点与其回归线之间关系的可视化图表。线性回归通过拟合一条线来建模自变量（输入变量）与因变量（输出变量）之间的线性关系。

## 线性回归的基本概念：
线性回归模型假设自变量$X$和因变量$Y$之间存在如下线性关系：
$$Y = \beta_0 + \beta_1 X + \epsilon$$
- $Y$：因变量（目标变量）
- $X$：自变量（特征变量）
- $\beta_0$：截距项
- $\beta_1$：斜率（回归系数）
- $\epsilon$：误差项

## 线性回归图的组成：
- 数据点：图中的每个点表示一个观察值，即一组自变量$X$和因变量$Y$的值。
- 回归线：表示最佳拟合直线，通常使用最小二乘法来计算，目的是最小化所有数据点到回归直线的垂直距离之和。

## 适用场景
- 可视化趋势：线性回归图能够清晰地展示自变量与因变量之间的线性关系
- 判断拟合效果：通过查看回归线与数据点的接近程度，可以直观判断模型的拟合效果
- 预测：回归直线可以用于预测新数据的因变量值

## sns.regplot()

- `x`: X轴变量名称
- `y`: Y轴变量名称
- `data`: DataFrame，包含用于绘图的数据。
- `scatter`: 布尔值，控制是否绘制散点。
- `order`: 指定回归线的多项式阶数。
- `ci`: 控制置信区间的大小。
- `fit_reg`: 控制是否绘制回归线。
- `color`: 设置绘图颜色。
- `scatter_kws`: 设置散点样式。
- `line_kws`: 设置回归线样式。

### 添加回归线方程

**numpy.polyfit(x, y, deg)**: 用于多项式拟合的函数
  - x: 拟合所使用的自变量数据
  - y: 拟合所使用的因变量数据
  - deg: 拟合次数，比如 deg=1 表示线性回归，deg=2 表示二次回归，deg=3 表示三次回归等

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# 加载数据
tips = sns.load_dataset('tips')

# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制线性回归图：展示总账单金额与小费之间的关系
sns.regplot(
    x='total_bill',      # 横坐标是 'total_bill' 列，表示总账单金额
    y='tip',             # 纵坐标是 'tip' 列，表示小费金额
    data=tips,           # 数据来源是 tips DataFrame
    order=3,             # 指定三次回归
    scatter_kws={        # 设置散点样式
        'color':'red',
        's':15,
        'alpha':0.5
    },
    line_kws={          # 设置回归线样式
        'color':'green',
        'lw':2,
        'ls':'--',
        'alpha':0.6
    }
)

# 提取数据：total_bill作为自变量，tip作为因变量
X = tips['total_bill'].values
y = tips['tip'].values

# 使用scipy的polyfit进行三次回归拟合
coefficients = np.polyfit(X, y, 3)  # 三次回归

# 创建回归方程的表达式
a, b, c, d = coefficients  # 获取三次回归的系数
equation = f"y = {a:.5f}x³ + {b:.5f}x² + {c:.5f}x + {d:.5f}"

# 在图表上添加回归方程文本，坐标为(2,9)
plt.text(2, 9, equation, fontsize=14, color='black')

# 设置图表的标题：fontsize表示字体大小
plt.title('Cubic Regression: Total Bill vs Tip', fontsize=16, fontweight='bold')

# 设置横轴标签
plt.xlabel('Total Bill ($)', fontsize=12)

# 设置纵轴标签
plt.ylabel('Tip ($)', fontsize=12)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### 计算R-squared值

**R²（R-squared）**，也称为决定系数，是回归分析中衡量模型拟合优度的重要指标。它表示模型解释了目标变量的变异程度。简而言之，R²告诉我们自变量（或特征）对因变量（或目标）变化的解释能力有多强。

**R² 的定义**
R² 是 1 减去“模型残差平方和”（RSS）与“总平方和”（TSS）的比值：
$$R^2 = 1 - \frac{RSS}{TSS}$$
其中：
- TSS（总平方和，Total Sum of Squares）：是目标变量的总变异性，等于目标变量的每个观测值与其平均值的差异的平方和。
$$TSS = \sum_{i=1}^n (y_i - \bar{y})^2$$
- RSS（残差平方和，Residual Sum of Squares）：是回归模型预测值与实际值之间的差异的平方和，即回归线无法解释的数据变异部分。
$$RSS = \sum_{i=1}^n (y_i - \hat{y}_i)^2$$
- $yₖ$：目标变量的实际观测值。 
- $ŷₖ$：回归模型预测的值。 
- $ȳ$：目标变量的均值。 

**R² 的值的范围**
- R² = 1：表示回归模型完美地拟合数据，所有点都落在回归线上，模型能够完全解释因变量的变异性。
- R² = 0：表示回归模型无法解释数据的任何变异，模型的拟合效果与直接使用目标变量的平均值一样。
- R² < 0：虽然 R² 最低为 0，但在某些特殊情况下，尤其是当模型非常糟糕时，R² 可能会变为负数。这通常表示模型预测的效果不如仅使用目标均值预测的效果。

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

# 加载数据
tips = sns.load_dataset('tips')

# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制线性回归图：展示总账单金额与小费之间的关系
sns.regplot(
    x='total_bill',      # 横坐标是 'total_bill' 列，表示总账单金额
    y='tip',             # 纵坐标是 'tip' 列，表示小费金额
    data=tips,           # 数据来源是 tips DataFrame
    order=3,             # 指定三次回归
    scatter_kws={        # 设置散点样式
        'color':'red',
        's':15,
        'alpha':0.5
    },
    line_kws={          # 设置回归线样式
        'color':'green',
        'lw':2,
        'ls':'--',
        'alpha':0.6
    }
)

# 提取数据：total_bill作为自变量，tip作为因变量
X = tips['total_bill'].values
y = tips['tip'].values

# 使用scipy的polyfit进行三次回归拟合
coefficients = np.polyfit(X, y, 3)  # 三次回归

# 创建回归方程的表达式
a, b, c, d = coefficients  # 获取三次回归的系数
equation = f"y = {a:.5f}x³ + {b:.5f}x² + {c:.5f}x + {d:.5f}"

# 在图表上添加回归方程文本，坐标为(2,9)
plt.text(2, 9, equation, fontsize=14, color='black')

# 计算回归方程的预测值
y_pred = np.polyval(coefficients, X)

# 计算 R² 值
r_squared = r2_score(y, y_pred)

# 在图表中添加 R² 值文本
plt.text(2, 8, f"R² = {r_squared:.4f}", fontsize=14, color='black')

# 设置图表的标题：fontsize表示字体大小
plt.title('Cubic Regression: Total Bill vs Tip', fontsize=16, fontweight='bold')

# 设置横轴标签
plt.xlabel('Total Bill ($)', fontsize=12)

# 设置纵轴标签
plt.ylabel('Tip ($)', fontsize=12)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### 计算RMSE值

**均方根误差（Root Mean Squared Error，RMSE）**是衡量回归模型拟合效果的常用指标。它是预测值与实际值之差的平方的平均数的平方根。
公式如下：
$$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - y_{\text{pred}_i})^2}$$
其中：
- $y_i$是实际值
- $y_{\text{pred}_i}$ 是预测值
- $n$ 是样本数量

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error

# 加载数据
tips = sns.load_dataset('tips')

# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制线性回归图：展示总账单金额与小费之间的关系
sns.regplot(
    x='total_bill',      # 横坐标是 'total_bill' 列，表示总账单金额
    y='tip',             # 纵坐标是 'tip' 列，表示小费金额
    data=tips,           # 数据来源是 tips DataFrame
    order=3,             # 指定三次回归
    scatter_kws={        # 设置散点样式
        'color':'red',
        's':15,
        'alpha':0.5
    },
    line_kws={          # 设置回归线样式
        'color':'green',
        'lw':2,
        'ls':'--',
        'alpha':0.6
    }
)

# 提取数据：total_bill作为自变量，tip作为因变量
X = tips['total_bill'].values
y = tips['tip'].values

# 使用scipy的polyfit进行三次回归拟合
coefficients = np.polyfit(X, y, 3)  # 三次回归

# 创建回归方程的表达式
a, b, c, d = coefficients  # 获取三次回归的系数
equation = f"y = {a:.5f}x³ + {b:.5f}x² + {c:.5f}x + {d:.5f}"

# 在图表上添加回归方程文本，坐标为(2,9)
plt.text(2, 9, equation, fontsize=14, color='black')

# 计算回归方程的预测值
y_pred = np.polyval(coefficients, X)

# 计算 R² 值
r_squared = r2_score(y, y_pred)

# 计算 RMSE 值
rmse = np.sqrt(mean_squared_error(y, y_pred))

# 在图表中添加 R² 和 RMSE 值文本
plt.text(2, 8, f"R² = {r_squared:.4f}", fontsize=14, color='black')
plt.text(2, 7, f"RMSE = {rmse:.4f}", fontsize=14, color='black')

# 设置图表的标题：fontsize表示字体大小
plt.title('Cubic Regression: Total Bill vs Tip', fontsize=16, fontweight='bold')

# 设置横轴标签
plt.xlabel('Total Bill ($)', fontsize=12)

# 设置纵轴标签
plt.ylabel('Tip ($)', fontsize=12)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```
