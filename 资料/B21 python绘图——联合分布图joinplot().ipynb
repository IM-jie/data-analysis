# joinplot

**sns.jointplot()** 是 Seaborn 中用于绘制联合分布图的函数。它能够显示两个变量之间的关系以及它们各自的分布情况。联合分布图不仅展示了两个变量之间的散点关系，还在图的边缘显示了每个变量的边际分布。

**适用场景**

- 探索变量之间的关系：用来查看两个连续变量之间的相关性、趋势和关系。
- 分析分布：通过边际分布图展示每个变量的分布情况，帮助理解数据的偏态、分布特征和集中趋势。
- 发现异常值或离群点：通过散点图、回归图等，帮助识别数据中的异常值或离群点，尤其在数据点密集时非常有用。
- 比较不同类别之间的关系：使用 hue 参数为不同类别的数据点上色，帮助比较不同类别在两个变量间的关系差异。

## sns.jointplot

- `x`: X轴变量名称
- `y`: Y轴变量名称
- `data`: DataFrame，包含用于绘图的数据。
- `hue`: 按类别变量分组绘图。
- `height`: 控制联合分布图大小。
- `kind`: 指定绘图类型。

## 联合分布图案例

penguins 数据集包含了 344 个企鹅的观察数据，每个企鹅都有以下 6 个特征：

- species: 企鹅的种类（目标变量）：Adelie、Chinstrap 和 Gentoo。
- island: 企鹅栖息的岛屿（分类变量）：Torgersen、Biscoe 和 Dream。
- bill_length_mm: 鸟喙的长度（单位：毫米，连续变量）
- bill_depth_mm: 鸟喙的深度（单位：毫米，连续变量）
- flipper_length_mm: 鳍的长度（单位：毫米，连续变量）
- body_mass_g: 企鹅的体重（单位：克，连续变量）
- sex: 性别

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载Seaborn内置的penguins数据集
penguins = sns.load_dataset("penguins")
penguins
```

### 基本联合分布图

**plt.suptitle()**: 用于设置图形（或图表）的总标题

```python
# 设置 Seaborn 风格
sns.set_style("darkgrid")

# 绘制联合分布图
sns.jointplot(
    x="flipper_length_mm",  # 横轴：鳍长
    y="body_mass_g",        # 纵轴：体重
    data=penguins,          # 数据集
)

# 设置标题
plt.suptitle("Joint Distribution: Flipper Length vs Body Mass", fontsize=16, fontweight='bold', y=1.02)

# 使用 plt.xlabel 和 plt.ylabel 来设置标签
plt.xlabel("Flipper Length (mm)", fontsize=12)
plt.ylabel("Body Mass (g)", fontsize=12)

# 增加网格线
plt.grid(True)

# 显示图表
plt.show()
```

### hue-按类别变量分组绘图

```python
# 设置 Seaborn 风格
sns.set_style("darkgrid")

# 绘制联合分布图，添加 hue 根据不同种类上色
sns.jointplot(
    x="flipper_length_mm",  # 横轴：鳍长
    y="body_mass_g",        # 纵轴：体重
    data=penguins,          # 数据集
    hue="species",          # 使用 hue 来区分不同的企鹅种类
)

# 设置标题
plt.suptitle("Joint Distribution: Flipper Length vs Body Mass by Species", fontsize=16, fontweight='bold', y=1.02)

# 使用 plt.xlabel 和 plt.ylabel 来设置标签
plt.xlabel("Flipper Length (mm)", fontsize=12)
plt.ylabel("Body Mass (g)", fontsize=12)

# 增加网格线
plt.grid(True)

# 显示图表
plt.show()
```

### height-控制联合分布图大小

```python
# 设置 Seaborn 风格
sns.set_style("darkgrid")

# 绘制联合分布图，添加 hue 根据不同种类上色
sns.jointplot(
    x="flipper_length_mm",  # 横轴：鳍长
    y="body_mass_g",        # 纵轴：体重
    data=penguins,          # 数据集
    hue="species",          # 使用 hue 来区分不同的企鹅种类
    height=6
)

# 设置标题
plt.suptitle("Joint Distribution: Flipper Length vs Body Mass by Species", fontsize=16, fontweight='bold', y=1.02)

# 使用 plt.xlabel 和 plt.ylabel 来设置标签
plt.xlabel("Flipper Length (mm)", fontsize=12)
plt.ylabel("Body Mass (g)", fontsize=12)

# 增加网格线
plt.grid(True)

# 显示图表
plt.show()
```
