# 均匀分布

## 基本概念

**均匀分布**是指所有取值在某个区间内出现的概率相等的概率分布。
- **核心特性**：在区间$[a,b]$内，任意子区间概率仅与其长度成正比，与位置无关。
- **类型**：
    - **连续均匀分布**：适用于连续变量（如测量误差、随机数生成）。
    - **离散均匀分布**：适用于离散变量（如掷骰子）。

## 概率密度函数

若随机变量$X \sim U(a, b)$，即在区间$[a, b]$上服从均匀分布，则其概率密度函数（PDF）为：
$$f(x) = \begin{cases} \frac{1}{b - a}, & a \leq x \leq b \\ 0, & \text{otherwise} \end{cases}$$

## 均值与方差

- **均值**：$\frac{a + b}{2} $ 
- **方差**：$\frac{(b - a)^2}{12}$

## 性质

- **对称性**：在区间中心对称。 
- **概率线性增长**：概率密度为常数
- **无偏性**：期望值正好在中点。 
- **独立性**：多次独立抽样互不影响。 
- **变换性质**：如果$X \sim U(0, 1)$，则$Y = a + (b - a)X \sim U(a, b)$

## 图形

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm
import seaborn as sns

# 设置风格
sns.set(style="darkgrid")

# 均匀分布参数
a, b = 2, 5
dist = uniform(loc=a, scale=b - a)

# 生成均匀分布样本
samples = dist.rvs(size=1000)

# 定义x轴范围
x = np.linspace(a - 1, b + 1, 500)

# PDF 曲线
pdf = dist.pdf(x)

# 标准正态分布曲线（用于对比中心极限定理）
x_norm = np.linspace(-4, 4, 500)
y_norm = norm.pdf(x_norm, 0, 1)

# 绘图
plt.figure(figsize=(10, 6))

# 绘制均匀分布 PDF
plt.plot(x, pdf, label=f'Uniform PDF U({a}, {b})', color='blue', linewidth=2)

# 样本直方图
plt.hist(samples, bins=30, density=True, alpha=0.5, color='skyblue', label='Sample Histogram')

# 设置图形属性
plt.title('Uniform Distribution with Sample Histogram and Normal Comparison', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

```
