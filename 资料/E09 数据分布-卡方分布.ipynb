# 卡方分布

## 基本概念

**卡方分布（Chi-square distribution）** 是一种连续概率分布，用于检验数据的变异程度或验证假设的适配性。它是一种非对称分布。

假设$X_1, X_2, \ldots, X_k$是独立同分布的标准正态分布随机变量，则卡方分布的统计量定义为： $$\chi^2 = \sum_{i=1}^k X_i^2$$ 其中，$k$是自由度（degrees of freedom, df）

## 卡方分布的概率密度函数

卡方分布的概率密度函数为：
$$f(x; k) = \frac{1}{2^{k/2} \Gamma(k/2)} x^{(k/2)-1} e^{-x/2}, \quad x > 0$$
其中：
$k$是自由度。
$\Gamma(k/2)$是伽马函数

## 卡方分布的均值与方差

- **均值**：$k$ 
- **方差**：$2k$

## 卡方分布的性质

- **非负性**：卡方分布的值总是非负的，即$\chi^2 \geq 0$
- **可加性**：如果$\chi_1^2 \sim \chi^2(k_1)$ 且$\chi_2^2 \sim \chi^2(k_2)$，那么$\chi_1^2 + \chi_2^2 \sim \chi^2(k_1 + k_2)$
- **近似正态性**：当自由度较小时，卡方分布偏态较强（右偏）；自由度增大时，逐渐趋近正态分布$\chi^2 \sim N(k, 2k)$

## 卡方分布的图形

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, norm

# 设置自由度
degrees_of_freedom = [1, 2, 5, 10]

# 创建x值范围
x = np.linspace(0, 20, 500)

# 绘制卡方分布和正态分布的对比
plt.figure(figsize=(10, 6))

# 绘制卡方分布
for df in degrees_of_freedom:
    plt.plot(x, chi2.pdf(x, df), label=f'Chi-Square (df = {df})')

# 绘制正态分布（用于比较高自由度下的近似性）
mean = 10  # 对应高自由度
std = np.sqrt(2 * mean)
plt.plot(x, norm.pdf(x, mean, std), 'g--', label='Normal Distribution', linewidth=2)

# 添加标题和标签
plt.title('Chi-Square Distribution with Normal Approximation', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend(title='Distributions', fontsize=12)
plt.grid(alpha=0.3)

# 显示图形
plt.show()
```

- 自由度较低时（例如$df=1$），分布偏态较强，右侧拖尾显著
- 随着自由度增加（例如$df=10$），分布逐渐趋于对称，接近正态分布的形状

# 与其他分布关系

## 卡方分布与正态分布的关系

### 卡方分布直接来源于正态分布的平方和

如果$X_1, X_2, \ldots, X_k$ 是独立同分布的标准正态分布随机变量（即$X_i \sim N(0, 1)$），那么：
$$\chi^2 = \sum_{i=1}^k X_i^2$$就服从自由度为$k$的卡方分布。 

**自由度**$k$表示独立正态变量的个数。因此，卡方分布实际上是若干正态分布的平方和。

### 近似正态性

当卡方分布的自由度$k$很大时，其分布逐渐接近正态分布，具体表现为：
$$\chi^2 \sim N(k, 2k)$$
均值：$k$。
方差：$2k$。

### 正态分布中的平方项直接构成卡方分布

正态分布中的平方项直接构成卡方分布。例如：
$$Z \sim N(0, 1)， 则Z^2 \sim \chi^2(1)$$
如果是多个独立正态分布平方和，则形成更高自由度的卡方分布。
