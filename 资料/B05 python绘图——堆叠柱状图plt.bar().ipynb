# 堆叠柱状图（Stacked Bar Chart）

堆叠柱状图（Stacked Bar Chart）是一种数据可视化图表，它通过将不同类别的数据值叠加在一起，形成多个堆叠的柱状图，来展示不同类别之间的比较和它们各自的总量。每个柱子代表一个总量，而柱子的不同颜色部分代表构成这个总量的各个部分的值。这种图表非常适合展示各个部分对整体的贡献以及随时间或不同条件下的变化。

**适用场景**
- **展示总量与部分量**：当需要展示一个总量由几个部分组成时，堆叠柱状图可以直观地表示出每个部分对总量的贡献。
- **比较不同类别**：可以比较不同类别（如不同产品、不同地区）在不同条件下（如不同时间段）的表现。
- **展示比例关系**：通过堆叠柱状图的不同颜色部分的长度，可以直观地看出各部分之间的比例关系。

## plt.bar
- `x`: 控制柱子的x轴位置。
- `height`: 控制柱子的高度。
- `bottom`: 控制柱子的底部起始位置。
- `color`: 设置柱子颜色。
- `hatch`: 设置填充图案。
- `label`: 设置图例标签。

## 堆叠柱状图案例

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 创建日期范围
dates = pd.date_range('20230101', periods=7)

# 创建数据集
data = {
    'Date': dates,
    'Product_A': np.random.randint(100, 500, len(dates)),
    'Product_B': np.random.randint(100, 500, len(dates)),
    'Product_C': np.random.randint(100, 500, len(dates)),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], len(dates))
}

# 将数据转换为DataFrame
df = pd.DataFrame(data)
df
```

### 基本堆叠柱状图

```python
# 设置Seaborn的风格
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(12, 7))  # 设置宽12英寸，高7英寸

# 初始化底部数组
bottom = np.zeros(len(df['Date']))

# 绘制堆叠柱状图
plt.bar(x=df['Date'], height=df['Product_A'], bottom=bottom)
bottom += df['Product_A']  # 更新底部，等价于bottom = bottom+df['Product_A']

plt.bar(x=df['Date'], height=df['Product_B'], bottom=bottom)
bottom += df['Product_B']  # 更新底部

plt.bar(x=df['Date'], height=df['Product_C'], bottom=bottom)
bottom += df['Product_C']  # 更新底部

# 设置图表标题和标签
plt.title('Total Sales by Product and Date', fontsize=16, fontweight='bold')  # 调整标题字体大小和粗细
plt.xlabel('Date', fontsize=12)  # 调整X轴标签字体大小
plt.ylabel('Sales', fontsize=12)  # 调整Y轴标签字体大小

# 添加网格线
plt.grid()

# 调整X轴日期标签的旋转角度，避免重叠
plt.xticks(rotation=45)

# 显示图表
plt.show()
```

### hatch设置填充图案

**填充图案清单**

| 符号 | 图案描述   |
|------|------------|
| '/'  | 斜线       |
| '`'  | 倒引号     |
| '-'  | 水平线     |
| '+'  | 交叉线     |
| 'x'  | 斜交叉线   |
| '.'  | 小点       |
| '*'  | 星形       |

```python
# 设置Seaborn的风格
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(12, 7))  # 设置宽12英寸，高7英寸

# 初始化底部数组
bottom = np.zeros(len(df['Date']))

# 绘制堆叠柱状图
plt.bar(x=df['Date'], height=df['Product_A'], bottom=bottom, color='blue', hatch='-')
bottom += df['Product_A']  # 更新底部

plt.bar(x=df['Date'], height=df['Product_B'], bottom=bottom, color='red', hatch='+')
bottom += df['Product_B']  # 更新底部

plt.bar(x=df['Date'], height=df['Product_C'], bottom=bottom, color='green', hatch='/')
bottom += df['Product_C']  # 更新底部

# 设置图表标题和标签
plt.title('Total Sales by Product and Date', fontsize=16, fontweight='bold')  # 调整标题字体大小和粗细
plt.xlabel('Date', fontsize=12)  # 调整X轴标签字体大小
plt.ylabel('Sales', fontsize=12)  # 调整Y轴标签字体大小

# 添加网格线
plt.grid()

# 调整X轴日期标签的旋转角度，避免重叠
plt.xticks(rotation=45)

# 显示图表
plt.show()
```

### label设置图例标签

```python
# 设置Seaborn的风格
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(12, 7))  # 设置宽12英寸，高7英寸

# 初始化底部数组
bottom = np.zeros(len(df['Date']))

# 绘制堆叠柱状图
plt.bar(x=df['Date'], height=df['Product_A'], bottom=bottom, color='blue', hatch='-', label='Product_A')
bottom += df['Product_A']  # 更新底部

plt.bar(x=df['Date'], height=df['Product_B'], bottom=bottom, color='red', hatch='+', label='Product_B')
bottom += df['Product_B']  # 更新底部

plt.bar(x=df['Date'], height=df['Product_C'], bottom=bottom, color='green', hatch='/', label='Product_C')
bottom += df['Product_C']  # 更新底部

# 设置图表标题和标签
plt.title('Total Sales by Product and Date', fontsize=16, fontweight='bold')  # 调整标题字体大小和粗细
plt.xlabel('Date', fontsize=12)  # 调整X轴标签字体大小
plt.ylabel('Sales', fontsize=12)  # 调整Y轴标签字体大小

# 添加网格线
plt.grid()

# 显示图例
plt.legend(title='Product')

# 调整X轴日期标签的旋转角度，避免重叠
plt.xticks(rotation=45)

# 显示图表
plt.show()
```
