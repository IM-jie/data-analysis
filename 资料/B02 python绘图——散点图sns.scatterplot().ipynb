# 散点图

散点图（Scatter Plot）是一种用于展示两个数值变量之间关系的图表。它通过在二维平面上绘制点来表示数据点，每个点的横坐标和纵坐标分别对应两个变量的数值。散点图可以展示变量之间的相关性、趋势、模式或异常值。

**适用场景**

- **探索变量关系**：分析两个连续变量之间的关系。
- **趋势分析**：识别数据中的趋势，如增长或下降趋势。
- **模式识别**：发现数据中的模式，如聚类或分组。
- **异常值分析**：检测数据中的异常值或离群点。

## sns.scatterplot
- `data`: 要绘制的数据集
- `x`: 用于绘制散点图的 x 轴数据。
- `y`: 用于绘制散点图的 y 轴数据。
- `hue`: 用于分组的变量，可以按分类变量显示不同的颜色。
- `style`: 控制点的形状（如圆圈、方形等）。

## 散点图案例

Iris鸢尾花数据集是一个经典的机器学习数据集，包含了三种鸢尾花（Setosa, Versicolor, Virginica）的不同特征，每个花种有 50 个样本，总共有 150 个样本
- sepal_length：花萼长度
- sepal_width：花萼宽度
- petal_length：花瓣长度
- petal_width：花瓣宽度
- species：花种

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载iris数据集
iris = sns.load_dataset('iris')
iris
```

### 基本散点图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制散点图，展示花瓣长度和花瓣宽度之间的关系
sns.scatterplot(data=iris, x='petal_length', y='petal_width')

# 设置图表标题和标签
plt.title('Petal Length vs. Petal Width by Species', fontsize=16, fontweight='bold')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### hue分组散点图

`hue='species'`：根据不同类别给每个点着色

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽10英寸，高6英寸

# 绘制散点图，展示花瓣长度和花瓣宽度之间的关系
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')

# 设置图表标题和标签
plt.title('Petal Length vs. Petal Width by Species', fontsize=16, fontweight='bold')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### style参数控制散点形状

`style='species'`：这个参数会根据每个花卉物种 (species) 来控制不同的标记形状

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽10英寸，高6英寸

# 绘制散点图，展示花瓣长度和花瓣宽度之间的关系
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species')

# 设置图表标题和标签
plt.title('Petal Length vs. Petal Width by Species', fontsize=16, fontweight='bold')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')


# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```
