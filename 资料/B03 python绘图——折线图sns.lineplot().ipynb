# 折线图

折线图（Line Chart）是一种用于展示数据随时间或有序类别变化趋势的图表。它通过直线段连接数据点，每个数据点表示一个具体的数值，而数据点之间的连线则显示了数值随时间或有序类别变化的趋势。

**适用场景**

- **时间序列分析**：展示随时间变化的数据，如月销售额、年降雨量等。
- **比较趋势**：比较不同类别或组之间的趋势，如不同产品的销售增长趋势。
- **显示变化速率**：展示数据变化的速率或加速度，如经济增长率。
- **长期趋势分析**：分析长期数据变化趋势，如人口增长、气候变化等。

## sns.lineplot
- `x`: 字符串或数组，X轴变量的名称或数据。
- `y`: 字符串或数组，Y轴变量的名称或数据。
- `data`: DataFrame，包含用于绘图的数据。
- `color`: 设置线条颜色，"red"、"green"、"blue"、"yellow"等
- `markers`: 字符串或列表，用于定义点的标记样式，'.'、'o' 、'*'、'x'、'+'、'D'等
- `linestyle`: 设置线条样式
    - '-' : 实线（默认）
    - '--' : 虚线
    - '-.' : 点划线
    - ':' : 点线
- `linewidth`: 设置线条宽度
- `errorbar`: 用于控制误差条的显示
    - None：不显示误差条。
    - 'ci'：显示置信区间（默认值）。
    - 'sd'：显示标准差。
    - 'se'：显示标准误。

## 折线图案例

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载flights数据集
flights = sns.load_dataset('flights')

flights
```

### 基本折线图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制折线图，展示1949年到1960年间每月乘客数量的变化
sns.lineplot(
    data=flights,       # 数据来源，`flights` 是DataFrame表格，包含了日期和乘客数量等信息
    x='year',           # 横坐标是 'year' 列，表示年份
    y='passengers'     # 纵坐标是 'passengers' 列，表示乘客数量
)

# 设置图表的标题：fontsize表示字体大小
plt.title('Monthly Airline Passengers 1949-1960', fontsize=16, fontweight='bold')
# 设置横轴标签
plt.xlabel('Year')
# 设置纵轴标签
plt.ylabel('Number of Passengers')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### errorbar='sd'

`sd`：显示标准差

标准差是用来衡量数据集的离散程度或波动性的一种度量，反映了数据点与数据平均值之间的差异。标准差公式如下：
$$\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2}$$
其中：
- $x_i$ 是每个数据点
- $\mu$ 是数据的均值
- $N$ 是数据点的总数

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 绘制折线图，展示1949年到1960年间每月乘客数量的变化
sns.lineplot(
    data=flights,       # 数据来源，`flights` 是DataFrame表格，包含了日期和乘客数量等信息
    x='year',           # 横坐标是 'year' 列，表示年份
    y='passengers',     # 纵坐标是 'passengers' 列，表示乘客数量
    color='blue',       # 设置折线的颜色为蓝色
    marker='o',         # 设置数据点的标记样式为圆圈 (o)
    linestyle='-',      # 设置折线为实线（默认是实线，'-' 表示）
    linewidth=2,        # 设置线条的宽度为2，线条更粗，易于观察
    errorbar='sd'       # 使用标准差
)

# 设置图表的标题：fontsize表示字体大小
plt.title('Monthly Airline Passengers 1949-1960', fontsize=16, fontweight='bold')
# 设置横轴标签
plt.xlabel('Year')
# 设置纵轴标签
plt.ylabel('Number of Passengers')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### errorbar='se'

`se`：显示标准误

标准误是样本均值的标准差，用于衡量样本均值的不确定性或估计误差，反映了如果多次抽样样本均值与真实总体均值之间的偏差可能是多少。标准误公式如下：
$$SE = \frac{\sigma}{\sqrt{n}} $$
其中：
- $\sigma$是样本标准差
- $n$是样本的大小

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 绘制折线图，展示1949年到1960年间每月乘客数量的变化
sns.lineplot(
    data=flights,       # 数据来源，`flights` 是DataFrame表格，包含了日期和乘客数量等信息
    x='year',           # 横坐标是 'year' 列，表示年份
    y='passengers',     # 纵坐标是 'passengers' 列，表示乘客数量
    color='blue',       # 设置折线的颜色为蓝色
    marker='o',         # 设置数据点的标记样式为圆圈 (o)
    linestyle='-',      # 设置折线为实线（默认是实线，'-' 表示）
    linewidth=2,        # 设置线条的宽度为2，线条更粗，易于观察
    errorbar='se'       # 使用标准误
)

# 设置图表的标题：fontsize表示字体大小
plt.title('Monthly Airline Passengers 1949-1960', fontsize=16, fontweight='bold')
# 设置横轴标签
plt.xlabel('Year')
# 设置纵轴标签
plt.ylabel('Number of Passengers')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

## 多条折线图

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载Seaborn内置的 'flights' 数据集
flights = sns.load_dataset("flights")

# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(10, 6))  # 设置宽10英寸，高6英寸

# 选择我们感兴趣的年份数据
data_1949 = flights[flights['year'] == 1949]
data_1955 = flights[flights['year'] == 1955]
data_1960 = flights[flights['year'] == 1960]

# 绘制1949年的折线图
sns.lineplot(
    data=data_1949,           # 1949年数据
    x='month',                 # 横坐标是 'month' 列，表示月份
    y='passengers',           # 纵坐标是 'passengers' 列，表示乘客数量
    label='1949',              # 设置图例标签为1949
    marker='o',               # 设置数据点的标记样式为圆圈 (o)
    linestyle='-',            # 设置折线为实线
    linewidth=2               # 设置线条宽度为2
)

# 绘制1955年的折线图
sns.lineplot(
    data=data_1955,           # 1955年数据
    x='month',                 # 横坐标是 'month' 列，表示月份
    y='passengers',           # 纵坐标是 'passengers' 列，表示乘客数量
    label='1955',              # 设置图例标签为1955
    marker='o',               # 设置数据点的标记样式为圆圈 (o)
    linestyle='-',            # 设置折线为实线
    linewidth=2               # 设置线条宽度为2
)

# 绘制1960年的折线图
sns.lineplot(
    data=data_1960,           # 1960年数据
    x='month',                 # 横坐标是 'month' 列，表示月份
    y='passengers',           # 纵坐标是 'passengers' 列，表示乘客数量
    label='1960',              # 设置图例标签为1960
    marker='o',               # 设置数据点的标记样式为圆圈 (o)
    linestyle='-',            # 设置折线为实线
    linewidth=2               # 设置线条宽度为2
)

# 设置图表的标题：fontsize表示字体大小
plt.title('Monthly Airline Passengers for 1949, 1955, and 1960', fontsize=16, fontweight='bold')
# 设置横轴标签
plt.xlabel('Month')
# 设置纵轴标签
plt.ylabel('Number of Passengers')

# 显示图例
plt.legend(title='Year')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```
