# 同一个图使用双坐标轴

**双坐标轴图（Dual Axis Plot）**是指在同一个图表中使用 两个不同的 Y 轴，这两个 Y 轴可以有不同的尺度或单位。通常，双坐标轴图用于展示两种具有不同量纲或范围的数据，并且这两种数据共享同一 X 轴。

## plt.twinx(): 创建第二个坐标轴

## 双坐标轴图案例

1967-2015年美国经济关键指标数据
- `date`: 日期
- `pce`: 个人消费支出
- `pop`: 美国总人口数
- `psavert`: 个人储蓄率
- `uempmed`: 失业时间中位数
- `unemploy`: 失业人数

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 加载数据
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/economics.csv")

# 提取日期进行转换
df['date'] = pd.to_datetime(df['date'])
df
```

### 折线图+折线图

```python
# 设置 Seaborn 样式
sns.set(style="darkgrid")

# 创建画布，宽为14，高为6
plt.figure(figsize=(14, 6))

# 绘制储蓄率数据（左Y轴）
sns.lineplot(x='date', y='psavert', data=df, color='red', label='Savings Rate (%)')

# 创建第二个坐标轴
plt.twinx()

# 绘制失业人数数据（右Y轴）
sns.lineplot(x='date', y='unemploy', data=df, color='blue', label='Unemployed (Thousands)')

# 设置标题
plt.title('Savings Rate and Unemployment over Time', fontsize=16, fontweight='bold')

# 设置图例
plt.legend()

# 显示图表
plt.show()
```

### 散点图+折线图

```python
# 设置 Seaborn 样式
sns.set(style="darkgrid")

# 创建画布，宽为14，高为6
plt.figure(figsize=(14, 6))

# 绘制储蓄率数据（左Y轴）
sns.scatterplot(x='date', y='psavert', data=df, color='red', label='Savings Rate (%)')

# 创建第二个坐标轴
plt.twinx()

# 绘制失业人数数据（右Y轴）
sns.lineplot(x='date', y='unemploy', data=df, color='blue', label='Unemployed (Thousands)')

# 设置标题
plt.title('Savings Rate and Unemployment over Time', fontsize=16, fontweight='bold')

# 设置图例
plt.legend()

# 显示图表
plt.show()
```
