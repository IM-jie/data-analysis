# 面积图

**面积图（Area Plot）**是一种通过填充曲线与轴之间的区域来可视化数据的图形。它是折线图的变种，将折线图下方的区域进行填充，通常用于展示某个时间段或不同类别数据的累计量变化

**适用场景**
- **展示累计值的变化**：适用于展示随时间或其他变量变化的累计值（如：总收入、总用户数等）。 
- **显示部分对整体的贡献**：通过堆叠多个数据系列，直观展示各部分对整体的相对贡献（如：不同部门收入在总收入中的占比）。 
- **比较多个数据系列的变化趋势**：适用于比较多个系列的数据随时间或其他变量的变化，观察它们之间的关系和差异。 
- **表现数据的上升/下降趋势**：直观地显示数据的增长或下降趋势，适合展示如股票价格、销售量等的波动变化。 

## plt.fill_between

- `x`: X轴变量数据。
- `y1`: Y轴变量数据。
- `data`: DataFrame，包含用于绘图的数据。
- `color`: 设置线条颜色，"red"、"green"、"blue"、"yellow"等
- `alpha`: 填充区域的透明度，取值范围是 0（完全透明）到 1（完全不透明）
- `label`: 图例标签，用于图例中标识该区域
- `hatch`: 为填充区域指定纹理样式

## 面积图案例

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 模拟数据：2000 到 2019 年的 GDP（单位：万亿美元）
data = {
    'Year': list(range(2000, 2020)),
    'USA': [10.28, 10.73, 11.39, 12.07, 12.62, 13.13, 13.66, 14.49, 15.17, 15.69, 16.18, 16.83, 17.42, 17.92, 18.60, 19.39, 20.49, 21.43, 22.37, 23.12],
    'China': [1.21, 1.46, 1.84, 2.26, 2.61, 3.27, 3.58, 4.53, 5.46, 6.01, 6.59, 7.18, 7.53, 8.21, 9.33, 10.49, 11.38, 12.22, 13.62, 14.34],
    'India': [0.45, 0.51, 0.58, 0.72, 0.85, 1.00, 1.19, 1.47, 1.75, 2.00, 2.30, 2.63, 3.10, 3.44, 3.74, 4.13, 4.55, 5.04, 5.41, 6.02],
    'Germany': [1.92, 2.13, 2.32, 2.49, 2.75, 2.97, 3.11, 3.37, 3.73, 4.06, 4.38, 4.64, 4.89, 5.08, 5.16, 5.38, 5.63, 5.73, 5.89, 5.99],
    'Japan': [4.89, 4.76, 4.92, 5.10, 5.12, 5.15, 5.27, 5.43, 5.52, 5.61, 5.68, 5.89, 5.89, 5.91, 5.87, 5.97, 5.92, 5.89, 5.77, 5.57]
}

# 创建 DataFrame
df = pd.DataFrame(data)
df
```

### 基本面积图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制面积图
plt.fill_between(x='Year', y1='China', data=df)

# 设置标题
plt.title('GDP Distribution by Country (2000-2019)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (Trillions USD)', fontsize=12)

# 显示图表
plt.show()
```

### 多个面积图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制面积图
plt.fill_between(x='Year', y1='USA', data=df)
plt.fill_between(x='Year', y1='China', data=df)
plt.fill_between(x='Year', y1='India', data=df)

# 设置标题
plt.title('GDP Distribution by Country (2000-2019)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (Trillions USD)', fontsize=12)

# 显示图表
plt.show()
```

### alpha控制透明度

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制面积图
plt.fill_between(x='Year', y1='USA', data=df, color='blue', alpha=0.4)
plt.fill_between(x='Year', y1='China', data=df, color='red', alpha=0.5)
plt.fill_between(x='Year', y1='India', data=df, color='green', alpha=0.6)

# 设置标题
plt.title('GDP Distribution by Country (2000-2019)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (Trillions USD)', fontsize=12)

# 显示图表
plt.show()
```

### label控制标签

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制面积图，并为不同国家指定颜色、透明度和标签
plt.fill_between(x='Year', y1='USA', data=df, color='blue', alpha=0.4, label='USA')
plt.fill_between(x='Year', y1='China', data=df, color='red', alpha=0.5, label='China')
plt.fill_between(x='Year', y1='India', data=df, color='green', alpha=0.6, label='India')

# 设置标题
plt.title('GDP Distribution by Country (2000-2019)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (Trillions USD)', fontsize=12)

# 显示图例
plt.legend(title='Country')

# 显示图表
plt.show()
```

### hatch设置填充图案

**填充图案清单**

| 符号 | 图案描述   |
|------|------------|
| '/'  | 斜线       |
| '`'  | 倒引号     |
| '-'  | 水平线     |
| '+'  | 交叉线     |
| 'x'  | 斜交叉线   |
| '.'  | 小点       |
| '*'  | 星形       |

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制面积图，并为不同国家指定颜色、透明度、标签和hatch纹理
plt.fill_between(x='Year', y1='USA', data=df, color='blue', alpha=0.4, label='USA', hatch='-')
plt.fill_between(x='Year', y1='China', data=df, color='red', alpha=0.5, label='China', hatch='+')
plt.fill_between(x='Year', y1='India', data=df, color='green', alpha=0.7, label='India', hatch='/')

# 设置标题
plt.title('GDP Distribution by Country (2000-2019)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (Trillions USD)', fontsize=12)

# 显示图例
plt.legend(title='Country')

# 显示图表
plt.show()
```
