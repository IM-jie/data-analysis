# 山脊图

**山脊图（Ridgeline Plot）**通过将每个类别的分布图（直方图或核密度估计图）堆叠在一起，形成像山脊一样的图形，从而实现对多个变量分进行可视化展示

## 适用场景

- 多分布比较：山脊图的最大优势是能够在同一个图表中展示多个数据集的分布情况。
- 数据探索：山脊图中的曲线一般是通过核密度估计（KDE）来绘制的。通过山脊图，你可以发现数据分布的差异、趋势、异常值等信息。比如，发现某个数据集比其他数据集的分布要更加集中，或者某些数据集之间有显著的重叠。

## joypy.joyplot

- `data`: 用于画图数据，DataFrame。
- `column`: 需要绘制的变量名称
- `by`: 按照某个字段对数据进行分组，这样就可以按组绘制不同的子图
- `fade`: 控制图形曲线是否随高度逐渐变得透明
- `overlap`: 控制重叠程度
- `color`: 设置颜色
- `hist&bins`: 是否将数据绘制为直方图

## 山脊图案例

**tips**: 餐饮小费数据集
- total_bill：账单金额
- tip：小费金额
- sex：性别（Male 或 Female）
- smoker：是否吸烟（Yes 或 No）
- day：星期几（Sun, Thu, Fri, Sat）
- time：就餐时间（Lunch 或 Dinner）
- size：就餐人数

```python
import joypy
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# 使用 seaborn 内置的 tips 数据集
tips = sns.load_dataset("tips")
tips
```

### 山脊图基本案例

```python
# 设置 Seaborn 风格
sns.set_style("darkgrid")

# 设置图像大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 使用 joypy 绘制山脊图
joypy.joyplot(
    data=tips,            # 传入整个 DataFrame
    column='total_bill',  # 需要绘制的变量名称
    by="day"              # 按 "day" 列分组
)

# 设置标题和标签
plt.title('Ridgeline Plot of Total Bill by Day', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### fade-控制图形曲线是否随高度逐渐变得透明

```python
# 设置 Seaborn 风格
sns.set_style("darkgrid")

# 设置图像大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 使用 joypy 绘制山脊图
joypy.joyplot(
    data=tips,            # 传入整个 DataFrame
    column='total_bill',  # 需要绘制的变量名称
    by="day",             # 按 "day" 列分组
    fade=True
)

# 设置标题和标签
plt.title('Ridgeline Plot of Total Bill by Day', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### overlap-控制重叠程度

默认为 1。值越小，数据系列之间的间距越大；值越大，数据系列之间的重叠程度越高

```python
# 设置 Seaborn 风格
sns.set_style("darkgrid")

# 设置图像大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 使用 joypy 绘制山脊图
joypy.joyplot(
    data=tips,            # 传入整个 DataFrame
    column='total_bill',  # 需要绘制的变量名称
    by="day",             # 按 "day" 列分组
    fade=True,
    overlap=1
)

# 设置标题和标签
plt.title('Ridgeline Plot of Total Bill by Day', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### hist&bins-是否将数据绘制为直方图

- hist: 
    - True: 绘制为直方图
    - False: 不绘制为直方图
- bins: 直方图的箱子数量，默认为 10

```python
# 设置 Seaborn 风格
sns.set_style("darkgrid")

# 设置图像大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 使用 joypy 绘制山脊图
joypy.joyplot(
    data=tips,            # 传入整个 DataFrame
    column='total_bill',  # 需要绘制的变量名称
    by="day",             # 按 "day" 列分组
    fade=True,
    overlap=1,
    hist=True,
    bins=100
)

# 设置标题和标签
plt.title('Ridgeline Plot of Total Bill by Day', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```
