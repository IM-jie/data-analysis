# 密度图

密度图（Density Plot）是一种用于展示连续变量概率分布的图形。它通过平滑的曲线来近似表示数据的概率密度函数，从而揭示数据的分布特征，如单峰、双峰、偏斜等。密度图常用于数据探索阶段，以直观地理解数据的形状和特征。

**适用场景**

- **单变量分析**：分析单个连续变量的分布情况。
- **双变量分析**：比较两个连续变量的分布差异。
- **模式识别**：识别数据中的异常值、聚集或间隙。

## sns.kdeplot:
- `data`：DataFrame 或数据集，指定数据来源。
- `x`：x轴数据名称。
- `y`：y轴数据名称。
- `fill`：是否填充曲线下方区域。
- `color`：字符串或颜色代码，指定曲线颜色。
- `alpha`：控制曲线透明度。
- `hue`：根据某个分类变量进行分组绘制不同颜色的密度曲线

## 密度图案例

**tips**: 餐饮小费数据集
- total_bill：账单金额
- tip：小费金额
- sex：性别（Male 或 Female）
- smoker：是否吸烟（Yes 或 No）
- day：星期几（Sun, Thu, Fri, Sat）
- time：就餐时间（Lunch 或 Dinner）
- size：就餐人数

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载Seaborn内置的数据集
tips = sns.load_dataset('tips')
tips
```

### 密度图基本案例

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制单变量密度图
sns.kdeplot(x='total_bill', data=tips)

# 设置标题和标签
plt.title('Density Plot of Total Bill', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### fill-是否填充曲线下方区域

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制单变量密度图
sns.kdeplot(x='total_bill', data=tips, fill=True)

# 设置标题和标签
plt.title('Density Plot of Total Bill', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### alpha-控制密度曲线的透明度，取值范围为 0 到 1。0 表示完全透明，1 表示完全不透明

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽10英寸，高6英寸

# 绘制单变量密度图
sns.kdeplot(x='total_bill', data=tips, fill=True, color="blue", alpha=0.1)

# 设置标题和标签
plt.title('Density Plot of Total Bill', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### hue-分组绘制多子图密度图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽10英寸，高6英寸

# 绘制单变量密度图
sns.kdeplot(x='total_bill', data=tips, fill=True, alpha=0.1, hue='sex')

# 设置标题和标签
plt.title('Density Plot of Total Bill', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### y-双变量密度图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽10英寸，高6英寸

# 绘制单变量密度图
sns.kdeplot(x='total_bill',y='tip', data=tips, fill=True, alpha=0.6, hue='sex')

# 设置标题和标签
plt.title('Density Plot of Total Bill', fontsize=16)
plt.xlabel('Total Bill (USD)', fontsize=14)
plt.ylabel('Tip (USD)', fontsize=14)

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```
