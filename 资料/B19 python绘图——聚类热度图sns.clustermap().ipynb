# 聚类热度图

**聚类热度图（Clustered Heatmap）**是一种结合了 热度图 和 聚类分析 的可视化方法，用于对聚类算法结果进行可视化展示。聚类算法是指对数据按照相似度进行划分，使得最相似的数据被分到一个蔟内。

**聚类热度图的组成**:
- 热度图（Heatmap）：通过色块的颜色来表示数据值的大小。
- 聚类（Clustering）：对数据进行聚类，将相似的行或列聚合在一起，通常使用层次聚类（Hierarchical Clustering）算法。
    - 行聚类：对行进行聚类。
    - 列聚类：对列进行聚类。

## sns.clustermap()

- `data`: DataFrame，包含用于绘图的数据。
- `cmap`: 选择颜色映射方案
- `method`: 选择聚类算法
- `metric`: 选择计算距离方法
- `annot`:控制是否在每个单元格中显示数值
- `fmt`: 选择数值显示格式

## 聚类热度图案例

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 创建一个示例数据框架（股票数据矩阵）
data = np.random.rand(10, 10)  # 10 个股票，10 天的价格数据
df = pd.DataFrame(data, columns=[f"Stock{i+1}" for i in range(10)],
                  index=[f"Day{i+1}" for i in range(10)])
df
```

### 基本聚类热度图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制聚类热度图
sns.clustermap(df)

# 设置标题
plt.title('Clustered Heatmap of Stock Prices', fontsize=16, fontweight='bold')

# 渲染、显示图表
plt.show()
```

### cmap控制颜色映射

**cmap清单**

**顺序色图（Sequential）**：适用于连续数据，通常用于表示数据的大小或趋势。
- Blues: 蓝色渐变色
- BuGn: 蓝绿色渐变色
- GnBu: 绿色到蓝色渐变色
- Oranges: 橙色渐变色
- Purples: 紫色渐变色
- YlGnBu: 黄色到绿色到蓝色渐变色
- YlOrRd: 黄色到橙色到红色渐变色

**离散色图（Diverging）**：适用于具有两极值的数据，通常用于表示数据的正负值或变化。
- PiYG: 绿色到紫色的渐变色
- RdBu: 红色到蓝色的渐变色
- RdYlBu: 红色到黄色到蓝色的渐变色
- coolwarm: 蓝色到红色的渐变色
- Spectral: 具有更丰富的色调的渐变色

**分类色图（Qualitative）**：适用于分类数据，颜色不一定有渐变，通常用不同的颜色代表不同的类别。
- Set1: 明亮的分类颜色
- Set2: 比 Set1 更柔和的分类颜色
- Set3: 包含更多颜色的分类色图
- Paired: 适用于一组配对的类别
- Pastel1: 淡色的分类颜色
- Pastel2: 另一组淡色的分类颜色

**灰度色图（Grayscale）**：适用于只有灰度值的情况，通常用于表示黑白图像或数据的强度。
- gray: 灰度色图
- Greys: 由黑色到白色的渐变

**特殊效果色图（Miscellaneous）**：各种不同风格的颜色映射，适用于不同的可视化效果。
- twilight: 用于环状数据，具有平滑的颜色过渡
- twilight_shifted: 类似于 twilight，但有一个平移
- hsv: 使用色相（hue）来映射数据，适用于周期性数据
- magma: 从黑色到黄色的渐变色，适用于暗背景图形
- inferno: 从黑色到黄色到红色的渐变色，适用于暗背景图形
- plasma: 明亮的色调，适用于展示高对比度的图表

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制聚类热度图
sns.clustermap(
    df,
    cmap="YlGnBu"            # 设置颜色映射
)

# 设置标题
plt.title('Clustered Heatmap of Stock Prices', fontsize=16, fontweight='bold')

# 渲染、显示图表
plt.show()
```

### method选择聚类算法

#### 常见method方法介绍

**'single'（最短距离法）**：每个簇之间的距离是簇内任意两个点之间的最短距离。此方法倾向于生成小的簇，适用于具有明显分离的小簇的情况。
- 优点：计算量小。
- 缺点：容易受到噪音和异常值的影响。

**'complete'（最远距离法）**：每个簇之间的距离是簇内任意两个点之间的最大距离。此方法倾向于生成较大的簇。
- 优点：适用于数据点较为分散的情况。
- 缺点：对噪音不够鲁棒。

**'average'（平均距离法）**：计算两个簇的平均距离作为簇之间的距离。
- 优点：相较于 single 和 complete，更加平衡，不容易受到异常值的影响。
- 缺点：计算量较大，适用于一般情况。

**'weighted'（加权距离法）**：类似于 average，但是对于不同簇的大小赋予不同的权重。
- 优点：能够更好地处理不同大小簇的数据。
- 缺点：适用于较为特殊的场景，计算量大。

**'centroid'（质心法）**：使用簇的质心（即簇内所有点的平均位置）之间的距离来计算簇与簇之间的距离。
- 优点：能捕捉到簇的中心位置，适用于数据点聚集性较强的情况。
- 缺点：无法处理簇内点的距离不一致的情况。

**'median'（中位数法）**：使用簇内点的中位数来计算簇与簇之间的距离。
- 优点：适用于数据相对分散的情况。
- 缺点：计算复杂，较为少见。

**'ward'（Ward方法，最小方差法）**：通过最小化合并后簇内的方差来选择簇。该方法在大多数情况下非常有效，且对噪声和异常值的鲁棒性较强。
- 优点：一般能得到较为平衡的聚类结果，适用于多种类型数据。
- 缺点：计算复杂度较高。

#### method使用案例

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制聚类热度图
sns.clustermap(
    df,
    cmap="YlGnBu",            # 设置颜色映射
    method="ward"           # 设置层次聚类的算法
)

# 设置标题
plt.title('Clustered Heatmap of Stock Prices', fontsize=16, fontweight='bold')

# 渲染、显示图表
plt.show()
```

### metric设置计算距离方法

#### 常见metric方法介绍

**'euclidean'（欧氏距离）**：计算两个点之间的直线距离。最常见的距离度量方式，适用于数值型数据。
- 公式： $d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}$
- 适用场景：适用于大多数连续型数据，尤其是点之间的空间距离度量。

**'manhattan'（曼哈顿距离）**：计算两个点之间在各个维度上差值的绝对值之和。也叫做“城市街区”距离，适用于网格状数据。
- 公式： $d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^n |x_i - y_i|$
- 适用场景：适用于具有格状结构的离散数据，例如城市地图或棋盘格。

**'cosine'（余弦相似度）**：衡量两个向量的方向差异，而不考虑它们的大小。适用于文本数据、TF-IDF向量等情况。
- 公式： $\text{cosine}(A, B) = \frac{A \cdot B}{||A|| \cdot ||B||}$
- 适用场景：适用于高维稀疏数据，常用于文本数据分析，衡量文档或词项之间的相似度。

**'hamming'（汉明距离）**：衡量两个字符串或二进制向量之间不同的位数。常用于分类标签或二进制数据。
- 公式： $d(\mathbf{x}, \mathbf{y}) = \text{number of differing bits}$
- 适用场景：适用于比较二进制数据（如 DNA 序列比对、标签预测问题）。

**'jaccard'（杰卡德相似度）**：衡量两个集合的相似性，计算两个集合交集与并集的比值。适用于集合数据或离散数据。
- 公式： $\text{Jaccard}(A, B) = \frac{|A \cap B|}{|A \cup B|}$
- 适用场景：适用于离散的集合数据，例如用户购买商品集合或社交网络中的相似性度量。

**'chebyshev'（切比雪夫距离）**：计算两个点在所有维度上的最大差异。适用于离散数据和非常规的度量标准。
- 公式： $d(\mathbf{x}, \mathbf{y}) = \max_i |x_i - y_i|$
- 适用场景：适用于需要考虑维度间最大差异的情况。

**'braycurtis'（布雷-柯蒂斯相似度）**：衡量两个样本间的差异，通常用于生态学和环境科学中，尤其是在物种分布数据中。
- 公式： $d(\mathbf{x}, \mathbf{y}) = \frac{\sum |x_i - y_i|}{\sum |x_i + y_i|}$
- 适用场景：适用于生态学数据，尤其是在物种组成比较时。

**'mahalanobis'（马氏距离）**：基于协方差矩阵计算的距离度量，考虑了数据的相关性。适用于多变量正态分布数据。
- 公式： $d(\mathbf{x}, \mathbf{y}) = \sqrt{(\mathbf{x} - \mathbf{y})^T S^{-1} (\mathbf{x} - \mathbf{y})}$
- 适用场景：适用于多维数据分析，特别是具有协方差结构的数据。

**'canberra'（坎贝尔距离）**：衡量两个点在各个维度上的差异，相比于曼哈顿距离，更加强调维度之间的差异。
- 公式： $d(\mathbf{x}, \mathbf{y}) = \sum \frac{|x_i - y_i|}{|x_i| + |y_i|}$
- 适用场景：适用于具有较大差异或零值的离散数据。

**'seuclidean'（标准欧几里得距离）**：欧几里得距离的标准化版本，适用于不同尺度的数值数据。
- 公式： $d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum \frac{(x_i - y_i)^2}{s_i^2}}$
- 适用场景：适用于尺度不一致的数据集。

#### metric使用案例

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制聚类热度图
sns.clustermap(
    df,
    cmap="YlGnBu",            # 设置颜色映射
    method="ward",           # 设置层次聚类的算法
    metric="euclidean"      # 设置计算距离的方法
)

# 设置标题
plt.title('Clustered Heatmap of Stock Prices', fontsize=16, fontweight='bold')

# 渲染、显示图表
plt.show()
```

### annot控制是否在每个单元格中显示数值

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制聚类热度图
sns.clustermap(
    df,
    cmap="YlGnBu",            # 设置颜色映射
    method="ward",           # 设置层次聚类的算法
    metric="euclidean",      # 设置计算距离的方法
    annot=True              # 在每个单元格中显示数值
)

# 设置标题
plt.title('Clustered Heatmap of Stock Prices', fontsize=16, fontweight='bold')

# 渲染、显示图表
plt.show()
```

### fmt控制显示格式

**浮动小数格式**：
- '.2f'：表示显示小数点后两位数字。例如，3.14159 会显示为 3.14。
- '.3f'：表示显示小数点后三位数字。例如，3.14159 会显示为 3.142。
- '.0f'：表示显示整数部分，不显示小数。例如，3.14159 会显示为 3。

**科学计数法**
- '{:.2e}'：使用科学计数法格式，保留两位小数。例如，0.00012345 会显示为 1.23e-04。
- '{:.2g}'：使用科学计数法格式，自动选择合适的精度，保留有效数字的两位。

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制聚类热度图
sns.clustermap(
    df,
    cmap="YlGnBu",            # 设置颜色映射
    method="ward",           # 设置层次聚类的算法
    metric="euclidean",      # 设置计算距离的方法
    annot=True,              # 在每个单元格中显示数值
    fmt='.2f'                # 控制显示格式
)

# 设置标题
plt.title('Clustered Heatmap of Stock Prices', fontsize=16, fontweight='bold')

# 渲染、显示图表
plt.show()
```
