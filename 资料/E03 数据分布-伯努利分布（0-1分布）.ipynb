# 伯努利分布

## 基本概念

**伯努利分布**又称为0-1分布，是一种最简单的离散型概率分布，描述**单次二元试验**的结果：
   - **试验场景**：只有两种可能结果的实验（如抛硬币、产品合格/不合格）。
   - **参数**：成功概率 p（失败概率 q=1−p）。
   - **随机变量**：$X∼Bernoulli(p)$，其中：
       - X=1（成功）
       - X=0（失败）

## 概率密度函数

假设随机变量$X \in \{0, 1\}$，则$X \sim \text{Bernoulli}(p)$，其中$p \in [0, 1]$是成功的概率。则其概率密度函数为

$$P(X = x) = 
\begin{cases} 
p & \text{if } x = 1, \\
1-p & \text{if } x = 0.
\end{cases}$$


## 均值与方差

- **均值**：$p$ 
- **方差**：$p(1-p)$

## 性质

- **二元性**：只能取0或1。
- **极值特性**：当 p=0.5，方差最大（$s^2=0.25$）。
- **可叠加性**：n次独立伯努利试验的和服从二项分布$Binomial(n,p)$。

## 图形

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import seaborn as sns

# 设置风格（参考原代码的darkgrid）
sns.set(style="darkgrid", palette="deep")

# 设置不同成功概率
probs = [0.2]
sample_size = 1000  # 每个分布的样本量

# 创建画布（保持10x6尺寸）
plt.figure(figsize=(10, 6))

# 为每个概率值绘制伯努利分布
for p in probs:
    # 生成伯努利随机变量
    data = bernoulli.rvs(p, size=sample_size)
    
    # 使用countplot展示频数分布（参考原代码的绘图风格）
    sns.countplot(x=data, 
                 label=f'Bernoulli (p={p})',
                 alpha=0.7,  # 保持透明度一致
                 edgecolor='black',  # 添加边框
                 linewidth=0.5)

# 美化图形（保持原代码风格）
plt.title('Bernoulli Distribution: Effect of Success Probability', 
          fontsize=16, pad=20)
plt.xlabel('Outcome', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks([0, 1], ['Failure (X=0)', 'Success (X=1)'])  # 明确标注分类

# 调整图例（位置和样式参考原代码）
plt.legend(title='Parameters', 
           title_fontsize=13,
           fontsize=12,
           framealpha=0.9)  # 半透明背景

# 网格和边框设置（保持原代码风格）
plt.grid(axis='y', alpha=0.3)  # 只显示y轴网格线
plt.gca().spines['top'].set_visible(False)  # 移除顶部边框
plt.gca().spines['right'].set_visible(False)  # 移除右侧边框

# 显示图形
plt.tight_layout()
plt.show()
```

# 伯努利分布与正态分布的关系

伯努利分布是离散的、二元的，而正态分布是连续的、对称的，它们之间存在间接联系：
- 多次独立伯努利试验 → 构成二项分布 
- 根据**中心极限定理**，在大样本下，二项分布近似正态分布： 
$$X \sim \text{Bin}(n, p) \approx N(np, np(1 - p)), \quad \text{当 } n \text{ 足够大}$$
