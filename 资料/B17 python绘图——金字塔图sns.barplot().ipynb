# 金字塔图（Population Pyramid）

**金字塔图（Pyramid Chart）**是一种用于展示层级关系或数据分布的可视化图表，形状类似金字塔，通常用于表示 人口结构、市场份额、销售数据、组织架构等。它适用于需要突出数据 比例、分布、层级关系 的场景。

**适用场景**
  
- **人口统计**：如 人口金字塔，比较不同年龄段的性别比例。
- **市场分析**：比较不同产品类别的销售占比。
- **组织结构**：用于企业的管理层级展示。
- **销售漏斗**：展示潜在客户到成交客户的转化情况。

## sns.barplot
- `data`: DataFrame，包含用于绘图的数据。
- `x`: X轴变量名称。
- `y`: Y轴变量名称。
- `color`: 设置颜色。
- `label`: 设置标签。
- `xticks`: 设置X轴刻度。

## 金字塔图案例

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 构造人口数据
data = {
    "Age Group": ["0-4", "5-9", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", 
                  "40-44", "45-49", "50-54", "55-59", "60-64", "65-69", "70-74", "75-79", "80-84", "85+"],
    "Male": [6.5, 6.3, 6.1, 6.0, 5.8, 5.5, 5.2, 5.0, 4.8, 4.5, 4.2, 3.8, 3.4, 3.0, 2.6, 2.2, 1.8, 1.5],
    "Female": [6.2, 6.1, 6.0, 6.2, 6.1, 5.7, 5.5, 5.3, 5.1, 4.9, 4.7, 4.4, 4.1, 3.8, 3.5, 3.1, 2.7, 2.3]
}

# 创建 DataFrame
df = pd.DataFrame(data)

# 重要！取负数，实现金字塔图
df["Female"] = -df["Female"]

df
```

### 基本金字塔图

```python
# 设置 Seaborn 样式，使得背景包含白色网格，增加图表的可读性
sns.set_style("whitegrid")

# 创建一个新的图形并设置图表的尺寸，宽度为 10，高度为 8
plt.figure(figsize=(10, 8))

# 绘制男性、女性人口数据的条形图，X 轴表示人口百分比，Y 轴表示年龄组
sns.barplot(data=df, x="Male", y="Age Group")   
sns.barplot(data=df, x="Female", y="Age Group")

# 设置 X 轴的标签、Y 轴的标签，字体大小为 12，粗体显示
plt.xlabel("Population Percentage (%)", fontsize=12, fontweight='bold')
plt.ylabel("Age Group", fontsize=12, fontweight='bold')
# 设置图表标题为 "Population Pyramid"，字体大小为 16，粗体显示
plt.title("Population Pyramid", fontsize=16, fontweight='bold')

# 显示图表，呈现最终的图形结果
plt.show()
```

### xticks设置 X 轴刻度

**plt.xticks(ticks, labels)**: 用于设置 X 轴的刻度
  - ticks: 刻度的位置
  - labels: 刻度的标签

```python
# 设置 Seaborn 样式，使得背景包含白色网格，增加图表的可读性
sns.set_style("whitegrid")

# 创建一个新的图形并设置图表的尺寸，宽度为 10，高度为 8
plt.figure(figsize=(10, 8))

# 绘制男性、女性人口数据的条形图，X 轴表示人口百分比，Y 轴表示年龄组
sns.barplot(data=df, x="Male", y="Age Group", label="Male")   
sns.barplot(data=df, x="Female", y="Age Group", label="Female")

# 设置 X 轴的标签、Y 轴的标签，字体大小为 12，粗体显示
plt.xlabel("Population Percentage (%)", fontsize=12, fontweight='bold')
plt.ylabel("Age Group", fontsize=12, fontweight='bold')
# 设置图表标题为 "Population Pyramid"，字体大小为 16，粗体显示
plt.title("Population Pyramid", fontsize=16, fontweight='bold')

# 设置 X 轴刻度
ticks=range(-10, 11, 2)
labels=[f"{abs(x)}%" for x in range(-10, 11, 2)]
plt.xticks(ticks, labels)

# 添加图例
plt.legend()

# 显示图表，呈现最终的图形结果
plt.show()
```

### 对每条数据添加数据标签

- **df["Male"][i]**：获取男性人口数据中第 i 个年龄组的百分比。
- **df["Male"][i] + 0.3**：将标签位置稍微偏移，使其不与条形重叠。+0.3 表示将标签向右移动 0.3 个单位。
- **i**：Y 轴的位置，即当前条形的年龄组的索引。
- **f"{df['Male'][i]}%"**：格式化字符串，显示男性人口的百分比。例如，如果男性人口为 6.5%，则显示 "6.5%"。
- **color="black"**：设置文本的颜色为黑色，以确保标签与背景对比明显。
- **va="center"**：设置垂直对齐方式为中心，使得文本垂直居中对齐，避免与条形图重叠。
- **fontsize=10**：设置字体大小为 10。
- **fontweight='bold'**：设置字体为粗体，增加可读性。

```python
# 设置 Seaborn 样式，使得背景包含白色网格，增加图表的可读性
sns.set_style("whitegrid")

# 创建一个新的图形并设置图表的尺寸，宽度为 10，高度为 8
plt.figure(figsize=(10, 8))

# 绘制男性、女性人口数据的条形图，X 轴表示人口百分比，Y 轴表示年龄组
sns.barplot(data=df, x="Male", y="Age Group", label="Male")   
sns.barplot(data=df, x="Female", y="Age Group", label="Female")

# 添加数据标签
for i in range(len(df)):
    plt.text(df["Male"][i] + 0.3, i, f"{df['Male'][i]}%", color="black", va="center", fontsize=10, fontweight='bold')
    plt.text(df["Female"][i] - 0.8, i, f"{abs(df['Female'][i])}%", color="black", va="center", fontsize=10, fontweight='bold')

# 设置 X 轴的标签、Y 轴的标签，字体大小为 12，粗体显示
plt.xlabel("Population Percentage (%)", fontsize=12, fontweight='bold')
plt.ylabel("Age Group", fontsize=12, fontweight='bold')
# 设置图表标题为 "Population Pyramid"，字体大小为 16，粗体显示
plt.title("Population Pyramid", fontsize=16, fontweight='bold')

# 设置 X 轴刻度
ticks=range(-10, 11, 2)
labels=[f"{abs(x)}%" for x in range(-10, 11, 2)]
plt.xticks(ticks, labels)

# 添加图例
plt.legend()

# 显示图表，呈现最终的图形结果
plt.show()
```
