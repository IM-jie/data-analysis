# t分布

## 基本概念

- t 分布 是一种对称、钟形的概率分布，主要用于以下场景：
    - 样本量较小（n < 30） 
    - 总体标准差未知 

- 它在小样本的统计推断中非常重要，如：
    - 单样本均值检验（t检验） 
    - 均值差异检验（成对样本、双样本） 
    - 构造置信区间（当 σ 未知） 

- $t$分布的分布记号：$t\sim t(\nu)$
    - 其中$\nu$是自由度（degrees of freedom），满足$\nu=n-1$。

## 概率密度函数

t 分布的概率密度函数为：
$$f(t) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu \pi} \, \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}$$

其中
- $\Gamma$是伽马函数（Gamma function）
- $\nu$是自由度。

## t分布的均值与方差

| 项目               | 值                                                                                                     |
|--------------------|-------------------------------------------------------------------------------------------------------|
| **均值 μ**         | 0（当自由度 $\nu > 1$）                                                                               |
| **方差 σ²**        | $\dfrac{\nu}{\nu - 2}$（当 $\nu > 2$）<br>∞（当 $1 < \nu \le 2$）<br>未定义（当 $\nu \le 1$）          |

## 性质

- **对称性**：关于 0 对称
- **厚尾性**：尾部比正态分布更厚（heavy tails）
- 自由度越大，越接近标准正态分布

## 图形

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm
import seaborn as sns
sns.set(style="darkgrid")

# 自由度设置
dfs = [1, 2, 5, 10, 30, 100]
x = np.linspace(-5, 5, 500)

# 绘图
plt.figure(figsize=(10, 6))

# 绘制不同自由度的 t 分布
for df in dfs:
    plt.plot(x, t.pdf(x, df), label=f't (df={df})')

# 绘制标准正态分布
plt.plot(x, norm.pdf(x), 'k--', label='Normal (μ=0, σ=1)', linewidth=2)

plt.title('t Distribution vs Standard Normal', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend(title='Distributions')
plt.grid(alpha=0.3)
plt.show()
```

# t分布与正态分布的关系

| 自由度 $\nu$  | t 分布形状          | 与正态分布比较       |
| ---------- | --------------- | ------------- |
| 小（如 1、2、3） | 较宽、尾部厚          | 更极端值、更分散      |
| 中（5～30）    | 逐渐接近正态分布        | 用于小样本估计       |
| 大（→∞）      | 逼近标准正态 $N(0,1)$ | t 分布极限是标准正态分布 |
