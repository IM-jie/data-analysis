# 正态分布

## 基本概念

正态分布（又称**高斯分布**）是统计学中最重要的连续概率分布之一，描述大量自然现象（如身高、测量误差、考试成绩等）的分布规律。其特点为：
- **对称性**：钟形曲线关于均值对称。
- **集中性**：数据集中在均值附近，远离均值的概率逐渐降低。
- **完全由参数决定**：仅需均值（μ）和标准差（σ）即可确定分布形态。
记作：
$$X \sim N(\mu, \sigma) $$

## 概率密度函数

正态分布的概率密度函数（PDF）为：
$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$
- **$\mu$**：均值，决定分布的中心位置。
- **$\sigma$**：标准差，决定曲线的“胖瘦”（数据离散程度）。
- **$e$**：自然常数（≈2.71828）。

## 正态分布的均值与标准差

- **均值**：$\mu$ 
- **标准差**：$\sigma$

## 关键参数与性质

| 参数     | 作用      | 影响                  |
| ------ | ------- | ------------------- |
| 均值（μ）  | 分布的中心位置 | 曲线沿x轴平移             |
| 标准差（σ） | 数据的离散程度 | σ越大，曲线越扁平；σ越小，曲线越陡峭 |

- **若数据服从正态分布，则满足68-95-99.7法则（经验法则）**：
    - 68%数据落在$\mu \pm \sigma$内，
    - 95%在$\mu \pm 2\sigma$内，
    - 99.7%在$\mu \pm 3\sigma$内。
- **线性变换**：若$X \sim N(\mu, \sigma^2)$，则$aX + b \sim N(a\mu + b, a^2\sigma^2)$。
- **中心极限定理**：当样本容量足够大时，不管原始变量的分布是什么，其样本均值的分布趋近于正态分布。

## 正态分布的图形

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

# 设置不同的均值和标准差组合
params = [
    (0, 1),
    (0, 2),
    (2, 1),
    (2, 0.5)
]

# 创建 x 值范围
x = np.linspace(-6, 6, 1000)

# 绘图
plt.figure(figsize=(10, 6))

# 绘制多个正态分布曲线
for mu, sigma in params:
    y = norm.pdf(x, mu, sigma)
    plt.plot(x, y, label=f'N(μ={mu}, σ={sigma})')

# 添加参考线（标准正态分布）
plt.plot(x, norm.pdf(x, 0, 1), 'k--', linewidth=2, label='Standard Normal (μ=0, σ=1)')

# 图形设置
plt.title('Normal Distributions with Different μ and σ', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend(title='Distributions', fontsize=12)
plt.grid(alpha=0.3)

# 显示图形
plt.show()
```

# 标准正态分布

## 基本概念

标准正态分布是一种特殊的正态分布，其均值（μ）为 0，标准差（σ）为 1，记作：
$$Z \sim N(0, 1) $$

## 概率密度函数

标准正态分布的概率密度函数为：
$$f(z) = \frac{1}{\sqrt{2\pi}} \exp\left( -\frac{z^2}{2} \right)$$

## 标准正态分布的均值与标准差

- **均值**：0
- **标准差**：1

## 性质

- **对称性**：
    - $\phi(-z) = \phi(z)$
    - $P(Z \leq -a) = 1 - P(Z \leq a)$
- **线性不变性**：若$Z \sim N(0, 1)$，则 $aZ + b \sim N(b, a^2)$

## 标准正态分布的图形

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

# 创建 x 值范围
x = np.linspace(-4, 4, 500)

# 计算标准正态分布的 PDF
pdf = norm.pdf(x, loc=0, scale=1)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', label='Standard Normal (μ=0, σ=1)', linewidth=2)

# 添加 μ±1σ, μ±2σ, μ±3σ 范围的可视化
plt.fill_between(x, pdf, where=(x >= -1) & (x <= 1), color='skyblue', alpha=0.4, label='μ ± 1σ')
plt.fill_between(x, pdf, where=(x >= -2) & (x <= 2), color='orange', alpha=0.2, label='μ ± 2σ')
plt.fill_between(x, pdf, where=(x >= -3) & (x <= 3), color='green', alpha=0.1, label='μ ± 3σ')

# 添加标题和标签
plt.title('Standard Normal Distribution (PDF)', fontsize=16)
plt.xlabel('z', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend(title='Range', fontsize=12)
plt.grid(alpha=0.3)

# 显示图形
plt.show()
```

# 标准正态分布与正态分布的关系

标准正态分布是一种均值（μ）为 0，标准差（σ）为 1的特殊的正态分布

## Z变换
任何正态分布$X \sim N(\mu, \sigma^2)$都可通过**Z变换**转换为标准正态分布:
$$Z = \frac{X - \mu}{\sigma}$$

## 逆变换
从标准正态分布还原到原正态分布
$$X = \mu + Z \cdot \sigma$$
