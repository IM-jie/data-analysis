# 直方图

直方图（Histogram）是一种统计图表，用于展示数据的分布情况。它通过将数据分组到连续的、不重叠的区间（或“箱子”）中，并以条形图的形式显示每个箱子中数据点的数量（或频率）来表示数据的分布。

**适用场景**

- **数据探索**：在数据分析的初步阶段，用于探索数据的分布特性。
- **质量控制**：在制造业中，用于分析产品尺寸等连续变量的质量分布。
- **市场调研**：在市场分析中，用于展示消费者年龄、收入等连续变量的分布情况。
- **金融风险控制**：在金融领域，用于分析投资回报率等连续变量的分布。

## sns.histplot
- `data`: 用于绘图的数据，DataFrame
- `x`: 指定要绘制直方图的变量名
- `y`: 如果提供了 y 参数，则绘制二维直方图
- `bins`: 指定直方图的箱数。可以是一个整数，指定箱的数量；也可以是一个序列，指定具体的箱边界。
- `kde`: 布尔值，决定是否在直方图上绘制概率密度估计曲线。
- `hue`: 分层变量，用于在直方图中创建多个子图，每个子图对应一个类别
- `stat`: 控制直方图y轴含义
- `cumulative`: 控制是否对数据进行累加

## 直方图案例

**tips**: 餐饮小费数据集
- total_bill：账单金额
- tip：小费金额
- sex：性别（Male 或 Female）
- smoker：是否吸烟（Yes 或 No）
- day：星期几（Sun, Thu, Fri, Sat）
- time：就餐时间（Lunch 或 Dinner）
- size：就餐人数

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载Seaborn内置的数据集
tips = sns.load_dataset('tips')
tips
```

### 基本直方图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(10, 6))  # 设置宽10英寸，高6英寸

# 绘制直方图
sns.histplot(data=tips, x='total_bill')

# 设置图表标题和标签
plt.title('Distribution of Total Bills', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Count')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### bins指定直方图的箱数

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(10, 6))  # 设置宽10英寸，高6英寸

# 绘制直方图
sns.histplot(data=tips, x='total_bill', bins=50)

# 设置图表标题和标签
plt.title('Distribution of Total Bills', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Count')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### kde-带概率密度曲线的直方图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(10, 6))  # 设置宽10英寸，高6英寸

# 绘制直方图
sns.histplot(data=tips, x='total_bill', bins=30, kde=True)

# 设置图表标题和标签
plt.title('Distribution of Total Bills', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Count')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### hue-多子图直方图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(10, 6))  # 设置宽10英寸，高6英寸

# 绘制直方图
sns.histplot(data=tips, x='total_bill', hue='sex' ,bins=30, kde=True)

# 设置图表标题和标签
plt.title('Distribution of Total Bills', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Count')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### stat-控制直方图y轴含义

- `count`：显示每个条形的计数（默认）。
- `density`：表示概率密度（条形高度归一化，使得面积总和为 1）。
- `probability`：用小数点表示的频率。
- `percent`：用百分数表示的频率。

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(10, 6))  # 设置宽10英寸，高6英寸

# 绘制直方图
sns.histplot(data=tips, x='total_bill', hue='sex' ,bins=30, kde=True, stat='percent')

# 设置图表标题和标签
plt.title('Distribution of Total Bills', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Percent(%)')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### cumulative-控制是否对数据进行累加

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(10, 6))  # 设置宽10英寸，高6英寸

# 绘制直方图
sns.histplot(data=tips, x='total_bill', bins=30, kde=True, stat='density',cumulative=True)

# 设置图表标题和标签
plt.title('Distribution of Total Bills', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Density')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```

### y-二维直方图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(10, 8))  # 设置宽10英寸，高6英寸

# 绘制直方图
sns.histplot(data=tips, x='total_bill', y='tip', hue='sex' ,bins=30, kde=True)

# 设置图表标题和标签
plt.title('Distribution of Total Bills', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip')

# 增加网格线
plt.grid(True)

# 渲染、显示图表
plt.show()
```
