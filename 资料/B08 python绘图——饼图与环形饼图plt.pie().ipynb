# 饼图

**饼图（Pie Chart）**是一种常用的数据可视化图表，通常用于显示各个部分与整体的比例关系。它的形状类似于一个圆形，被分成若干个扇形区域，每个区域代表整体中某一部分的比例。饼图的优势在于直观地显示不同部分所占的比例或百分比，使得观察者能够一眼看出各部分的大小对比。

**适用场景：**

- **展示比例**：当需要展示各部分占整体的比例时，如市场份额、人口分布等。
- **比较部分**：比较不同类别的相对大小，如不同产品的销售占比。
- **简化数据**：当数据集较为简单，分类不多时，饼图可以清晰地展示数据。

## plt.pie
- `x`：饼图中每个扇区的数值，通常是一个数字序列
- `labels`：标签，通常是一个与 x 数量相同的序列，用于标识每个扇区
- `colors`：指定每个扇区的颜色
- `autopct`：显示数值/百分比的格式
- `startangle`：控制绘图放入起始角度
- `explode`：用于突出显示某一部分，可以传入一个与 x 长度相同的列表，指定某部分的突出程度
- `shadow`：是否添加阴影效果
- `hatch`：设置填充图案
- `wedgeprops`：设置环形饼图
- `textprops`：设置显示字体属性
- `plt.legend`：设置图例

## 饼图案例

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 创建数据
data = {
    'Category': ['Electronics', 'Clothing', 'Home Goods', 'Beauty', 'Food', 'Sports Goods'],
    'Sales': [350, 200, 150, 100, 75, 125]
}

# 创建 DataFrame
df = pd.DataFrame(data)
df
```

### 基本饼图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图
plt.pie(x=df['Sales'], labels=df['Category'])

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### 颜色参数colors

#### 常见颜色调色板palette介绍

**Qualitative Palettes**：适合离散的类别数据，颜色彼此之间没有顺序或关联。
 
- 'deep'：默认的调色板，适合大部分情况。
- 'muted'：颜色较为柔和，适合多个类别的对比。
- 'pastel'：较为柔和的颜色，视觉上不那么刺眼。
- 'dark'：较为深沉的颜色，适合正式的图表。
- 'colorblind'：设计考虑到色盲友好，适合需要考虑视觉障碍的用户。
- 'Set1', 'Set2', 'Set3'：非常适合多个类别的颜色分配。
- 'Paired'：适合两两配对的颜色。
- 'Accent'：色调较鲜艳，适合强调特定类别。

**Sequential Palettes**：适合连续数据的颜色映射，从浅色到深色表示数据的从小到大的变化。
 
- 'Blues'：从浅蓝色到深蓝色。
- 'Greens'：从浅绿色到深绿色。
- 'Reds'：从浅红色到深红色。
- 'Purples'：从浅紫色到深紫色。
- 'YlGn'、'YlGnBu' 等：从黄色到绿色或蓝色。
- 'Oranges'、'YlOrBr'：从浅橙色到深橙色。
- 'BuGn'、'BuPu'：从浅蓝色到紫色等。

**Diverging Palettes**：适合表示有中值的连续数据，颜色从两端扩展到中间。
 
- 'RdBu'：从红色到蓝色，适合表示负值和正值。
- 'RdYlBu'：从红色到黄色再到蓝色，适合显示中立值。
- 'Spectral'：从紫色到蓝色再到红色，适合强烈的对比。

**Cyclic Palettes**：适合具有周期性的数据，颜色从一个极端到另一个极端，然后回到第一个极端。
 
- 'twilight'、'twilight_shifted'：常用于时间序列数据。

#### 不同colors参数案例

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图
plt.pie(x=df['Sales'], labels=df['Category'], colors=sns.color_palette("Set1"))

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### autopct显示数值/百分比

#### 仅显示百分比

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(x=df['Sales'], labels=df['Category'], colors=sns.color_palette("Set2"), autopct='%.2f%%')

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

#### 显示原始数值与百分比

```python
# 定义一个函数，显示原始数值和百分比
def display_values(pct, allvals):
    # pct表示百分比占比
    absolute = round(pct / 100.*sum(allvals), 0)  # 计算原始数值
    return f'{absolute:.0f} ({pct:.2f}%)'
```

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(
    x=df['Sales'], 
    labels=df['Category'], 
    colors=sns.color_palette("Set2"), 
    autopct=lambda pct: display_values(pct, df['Sales'])  # 显示原始数值和百分比
)

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### startangle控制起始角度

默认为从 x 轴正方向逆时针画起，如设定 startangle=90 则从 y 轴正方向画起

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(
    x=df['Sales'], 
    labels=df['Category'], 
    colors=sns.color_palette("Set2"), 
    autopct='%.2f%%',   # 显示百分比，保留1位小数
    startangle=140      # 设置起始角度为140度
)

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### explode突出显示某一部分

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(
    x=df['Sales'], 
    labels=df['Category'], 
    colors=sns.color_palette("Set2"), 
    autopct='%.2f%%',   # 显示百分比，保留1位小数
    startangle=140,      # 设置起始角度为140度
    explode=[0.1, 0.2, 0, 0, 0, 0]    # 设置突出显示效果
)

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### shadow设置阴影效果

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(
    x=df['Sales'], 
    labels=df['Category'], 
    colors=sns.color_palette("Set2"), 
    autopct='%.2f%%',   # 显示百分比，保留1位小数
    startangle=140,      # 设置起始角度为140度
    explode=[0.1, 0.2, 0, 0, 0, 0],    # 设置突出显示效果
    shadow=True         # 添加阴影效果
)

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### hatch设置填充图案

- /：斜线填充
- \\：反斜线填充
- |：竖直线填充
- -：水平线填充
- +：十字填充
- .：点状填充

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(
    x=df['Sales'], 
    labels=df['Category'], 
    colors=sns.color_palette("Set2"), 
    autopct='%.2f%%',   # 显示百分比，保留1位小数
    startangle=140,      # 设置起始角度为140度
    explode=[0.1, 0.2, 0, 0, 0, 0],    # 设置突出显示效果
    shadow=True,         # 添加阴影效果
    hatch=['/', '\\', '|', '-', '+', '.']   #设置填充图案
)

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### wedgeprops+width设置环形饼图

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 6))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(
    x=df['Sales'], 
    labels=df['Category'], 
    colors=sns.color_palette("Set2"), 
    autopct='%.2f%%',   # 显示百分比，保留1位小数
    startangle=140,      # 设置起始角度为140度
    explode=[0.1, 0.2, 0, 0, 0, 0],    # 设置突出显示效果
    shadow=True,         # 添加阴影效果
    wedgeprops={'width': 0.5},  # 环形饼图
    hatch=['/', '\\', '|', '-', '+', '.']   #设置填充图案
)

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### textprops设置显示字体属性

- fontsize: 字体大小
- fontweight: 字体粗细
    - 'normal': 正常字体
    - 'bold': 加粗
    - 'light': 轻体
- color: 字体颜色
- fontstyle: 字体样式
    - 'normal': 常规字体
    - 'italic': 斜体
    - 'oblique': 倾斜字体
- fontname: 字体名称。常见的字体名称有 'Arial'、'Helvetica'、'Times New Roman' 等
- va: 字体垂直对齐方式
    - 'top': 顶部对齐
    - 'center': 垂直居中
    - 'bottom': 底部对齐
- ha: 水平对齐方式
    - 'left': 左对齐
    - 'center': 居中对齐
    - 'right': 右对齐
- rotation: 文字旋转角度

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 8))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(
    x=df['Sales'], 
    labels=df['Category'], 
    colors=sns.color_palette("Set2"), 
    autopct='%.2f%%',   # 显示百分比，保留1位小数
    startangle=0,      # 设置起始角度为140度
    explode=[0.1, 0.2, 0, 0, 0, 0],    # 设置突出显示效果
    shadow=True,         # 添加阴影效果
    wedgeprops={'width': 0.5},  # 环形饼图
    hatch=['/', '\\', '|', '-', '+', '.'],   #设置填充图案
    textprops={
        'fontsize': 12,  # 设置字体大小
        'fontweight': 'bold',  # 设置字体加粗
        'color': 'black',  # 设置字体颜色为黑色
        'verticalalignment': 'center',  # 垂直居中
        'horizontalalignment': 'center',  # 水平居中
    }
)

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 显示图表
plt.show()
```

### plt.legend设置图例

- labels: 图例中每个元素的标签
- 图例的位置: 
    - 'best'：自动选择最佳位置（默认值）。matplotlib 会自动选择图例最不遮挡数据的位置。
    - 'upper right'：右上角。
    - 'upper left'：左上角。
    - 'lower left'：左下角。
    - 'lower right'：右下角。
    - 'right'：右侧中间。
    - 'center left'：左侧中间。
    - 'center right'：右侧中间。
    - 'lower center'：下方居中。
    - 'upper center'：上方居中。
    - 'center'：图表的中心位置
- fontsize: 图例字体大小
- title: 图例标题
- frameon: 控制图例边框是否显示
- shadow: 控制图例是否有阴影效果
- ncol: 设置图例中列的数目

```python
# 设置Seaborn的风格和颜色调色板
sns.set_style("darkgrid")

# 设置图片大小
plt.figure(figsize=(8, 8))  # 设置宽8英寸，高6英寸

# 绘制饼图，%.2f%%表示保留两位小数
plt.pie(
    x=df['Sales'], 
    labels=df['Category'], 
    colors=sns.color_palette("Set2"), 
    autopct='%.2f%%',   # 显示百分比，保留1位小数
    startangle=0,      # 设置起始角度为140度
    explode=[0.1, 0.2, 0, 0, 0, 0],    # 设置突出显示效果
    shadow=True,         # 添加阴影效果
    wedgeprops={'width': 0.5},  # 环形饼图
    hatch=['/', '\\', '|', '-', '+', '.'],   #设置填充图案
    textprops={
        'fontsize': 12,  # 设置字体大小
        'fontweight': 'bold',  # 设置字体加粗
        'color': 'black',  # 设置字体颜色为黑色
        'verticalalignment': 'center',  # 垂直居中
        'horizontalalignment': 'center',  # 水平居中
    }
)

# 设置标题和标签
plt.title('Sales Distribution by Category', fontsize=16)

# 添加图例，显示各个类别的颜色映射
plt.legend(labels=df['Category'], title='Category', loc='lower right', fontsize=8, shadow=True)

# 显示图表
plt.show()
```
